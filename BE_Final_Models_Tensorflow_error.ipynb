```python
# ====== Imports ======
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import os

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix

import tensorflow as tf
from tensorflow.keras import layers, models

# ====== Reproducibility ======
np.random.seed(42)
tf.random.set_seed(42)

# ====== Load Dataset ======
print("[INFO] Loading dataset...")
df = pd.read_csv("new_network_train.csv")
X = df.drop(columns=["ProtocolName"])
y = df["ProtocolName"].astype(np.int64)

# ====== Standardization ======
print("[INFO] Standardizing data...")
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, "scaler.pkl")
print("[INFO] Scaler saved as 'scaler.pkl'")

# ====== Train/Test Split ======
print("[INFO] Splitting dataset...")
X_train, X_val, y_train, y_val = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

# ====== Reshape for Conv1D ======
X_train = X_train[..., np.newaxis]
X_val = X_val[..., np.newaxis]

# ====== Parameters ======
input_dim = X_train.shape[1]
num_classes = len(np.unique(y))

# ====== Build CNN Model ======
def build_cnn_model(input_dim, num_classes):
    print("[INFO] Building CNN model...")
    model = models.Sequential([
        layers.Input(shape=(input_dim, 1)),
        layers.Conv1D(32, kernel_size=3, padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling1D(pool_size=2),

        layers.Conv1D(64, kernel_size=3, padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling1D(pool_size=2),

        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    model.summary()
    return model

# ====== Plot Training History ======
def plot_history(history):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Acc')
    plt.plot(history.history['val_accuracy'], label='Val Acc')
    plt.title('Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title('Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

# ====== Plot Confusion Matrix ======
def plot_confusion_matrix(y_true, y_pred, class_names):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=class_names,
                yticklabels=class_names)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title("Confusion Matrix")
    plt.tight_layout()
    plt.show()

# ====== Training & Evaluation Pipeline ======
start_time = time.time()

model = build_cnn_model(input_dim, num_classes)

print("[INFO] Starting model training...")
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=20,
    batch_size=64
)

print("[INFO] Saving model as 'cnn_model.h5'...")
model.save("cnn_model.keras")

plot_history(history)

print("[INFO] Evaluating model...")
loss, acc = model.evaluate(X_val, y_val, verbose=0)
print(f"[RESULT] Validation Accuracy: {acc:.4f}, Loss: {loss:.4f}")

print("[INFO] Generating predictions...")
y_pred_probs = model.predict(X_val)
y_pred_classes = np.argmax(y_pred_probs, axis=1)

class_names = [f"Class {i}" for i in range(num_classes)]

print("[INFO] Classification Report:")
print(classification_report(y_val, y_pred_classes, target_names=class_names))

print("[INFO] Confusion Matrix:")
plot_confusion_matrix(y_val, y_pred_classes, class_names)

end_time = time.time()
elapsed_time = end_time - start_time
print(f"[INFO] Total time taken: {elapsed_time:.2f} seconds")
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import joblib
import time
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns

# === Load Data ===
df = pd.read_csv("new_network_train.csv")
X = df.drop(columns=["ProtocolName"]).values
y = df["ProtocolName"].astype(np.int64).values

# === Label Map ===
label_map = {
    0: 'AMAZON', 1: 'CLOUDFLARE', 2: 'DROPBOX', 3: 'FACEBOOK',
    4: 'GMAIL', 5: 'GOOGLE', 6: 'HTTP', 7: 'HTTP_CONNECT', 8: 'HTTP_PROXY',
    9: 'MICROSOFT', 10: 'MSN', 11: 'SKYPE', 12: 'SSL', 13: 'TWITTER',
    14: 'WINDOWS_UPDATE', 15: 'YAHOO', 16: 'YOUTUBE'
}

# === Preprocessing ===
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, "scaler.pkl")

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, stratify=y, random_state=42)

# === Model Parameters ===
input_dim = X.shape[1]
latent_dim = 64
num_classes = len(np.unique(y))

# === Model Definitions ===
def build_encoder(input_dim, latent_dim):
    return tf.keras.Sequential([
        tf.keras.layers.Dense(512, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(256, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(latent_dim, activation='relu'),
    ], name="encoder")

def build_decoder(latent_dim, output_dim):
    return tf.keras.Sequential([
        tf.keras.layers.Dense(256, activation='relu'),
        tf.keras.layers.Dense(512, activation='relu'),
        tf.keras.layers.Dense(output_dim, activation='sigmoid'),
    ], name="decoder")

def build_classifier(latent_dim, num_classes):
    return tf.keras.Sequential([
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(num_classes, activation='softmax'),
    ], name="classifier")

# === Instantiate Models ===
encoder = build_encoder(input_dim, latent_dim)
decoder = build_decoder(latent_dim, input_dim)
classifier = build_classifier(latent_dim, num_classes)

# === Inputs ===
input_layer = tf.keras.Input(shape=(input_dim,))
encoded = encoder(input_layer)
decoded = decoder(encoded)
classified = classifier(encoded)

model = tf.keras.Model(inputs=input_layer, outputs=[decoded, classified])

# === Compile ===
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss=["mse", "sparse_categorical_crossentropy"],
    metrics=[None, "accuracy"]
)

# === Train Model ===
start_time = time.time()
print("\n🔁 Starting Joint AE + Classifier Training...\n")

history = model.fit(
    X_train, [X_train, y_train],
    validation_split=0.1,
    epochs=30,
    batch_size=128,
    verbose=1
)
end_time = time.time()

print(f"\n⏱️ Total Training Time: {(end_time - start_time):.2f} seconds")

# === Plot Classifier Accuracy ===
plt.plot(history.history['classifier_accuracy'], label="Train Acc")
plt.plot(history.history['val_classifier_accuracy'], label="Val Acc")
plt.title("Classifier Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()

# === Evaluation ===
print("\n✅ Evaluating on Test Data")
encoded_test = encoder.predict(X_test)
y_probs = classifier.predict(encoded_test)
y_pred = np.argmax(y_probs, axis=1)

acc = accuracy_score(y_test, y_pred)
print(f"\n✅ Final Test Accuracy: {acc:.4f}")

print("\n📊 Classification Report:\n")
print(classification_report(y_test, y_pred, target_names=[label_map[i] for i in sorted(label_map)]))

# === Confusion Matrix ===
plt.figure(figsize=(12, 10))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=[label_map[i] for i in sorted(label_map)],
            yticklabels=[label_map[i] for i in sorted(label_map)])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# === Save Models ===
# === Save Models in `.keras` format ===
encoder.save("encoder_joint_tf.keras")
decoder.save("decoder_joint_tf.keras")
classifier.save("classifier_joint_tf.keras")

print("\n💾 Models saved as:")
print("  - encoder_joint_tf.keras")
print("  - decoder_joint_tf.keras")
print("  - classifier_joint_tf.keras")

```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score
import joblib
import os

# === Label Map ===
label_map = {
    0: 'AMAZON', 1: 'CLOUDFLARE', 2: 'DROPBOX', 3: 'FACEBOOK',
    4: 'GMAIL', 5: 'GOOGLE', 6: 'HTTP', 7: 'HTTP_CONNECT', 8: 'HTTP_PROXY',
    9: 'MICROSOFT', 10: 'MSN', 11: 'SKYPE', 12: 'SSL', 13: 'TWITTER',
    14: 'WINDOWS_UPDATE', 15: 'YAHOO', 16: 'YOUTUBE'
}

# === Load and Preprocess Data ===
df = pd.read_csv("new_network_train.csv")
X = df.drop(columns=["ProtocolName"]).values
y = df["ProtocolName"].astype(np.int64).values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, "scaler.pkl")

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, stratify=y)
num_classes = len(np.unique(y))
input_dim = X.shape[1]
latent_dim = 100
label_smooth_real = 0.9

# === TensorFlow Dataset ===
batch_size = 128
train_dataset = tf.data.Dataset.from_tensor_slices((X_train, y_train)).shuffle(10000).batch(batch_size)
test_dataset = tf.data.Dataset.from_tensor_slices((X_test, y_test)).batch(batch_size)

# === One-hot encoding helper ===
def one_hot(labels, num_classes=num_classes):
    return tf.one_hot(labels, num_classes)

# === Focal Loss ===
def focal_loss(gamma=2.0):
    def loss_fn(y_true, y_pred):
        y_true_oh = tf.one_hot(y_true, depth=num_classes)
        y_pred = tf.nn.softmax(y_pred)
        epsilon = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)
        cross_entropy = -y_true_oh * tf.math.log(y_pred)
        loss = tf.reduce_sum((1 - y_pred) ** gamma * cross_entropy, axis=1)
        return tf.reduce_mean(loss)
    return loss_fn

# === Generator ===
def build_generator():
    noise = tf.keras.Input(shape=(latent_dim,))
    label = tf.keras.Input(shape=(num_classes,))
    x = tf.keras.layers.Concatenate()([noise, label])
    x = tf.keras.layers.Dense(256)(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)
    x = tf.keras.layers.Dense(512)(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)
    x = tf.keras.layers.Dense(input_dim)(x)
    return tf.keras.Model([noise, label], x, name="Generator")

# === Discriminator ===
def build_discriminator():
    data_input = tf.keras.Input(shape=(input_dim,))
    label_input = tf.keras.Input(shape=(num_classes,))
    x = tf.keras.layers.Concatenate()([data_input, label_input])
    x = tf.keras.layers.Dense(512)(x)
    x = tf.keras.layers.LeakyReLU(0.2)(x)
    x = tf.keras.layers.Dropout(0.3)(x)
    x = tf.keras.layers.Dense(256)(x)
    x = tf.keras.layers.LeakyReLU(0.2)(x)
    x = tf.keras.layers.Dropout(0.3)(x)
    x = tf.keras.layers.Dense(1, activation="sigmoid")(x)
    return tf.keras.Model([data_input, label_input], x, name="Discriminator")

# === CNN Classifier ===
def build_cnn_classifier():
    x_input = tf.keras.Input(shape=(input_dim,))
    x = tf.keras.layers.Reshape((input_dim, 1))(x_input)
    x = tf.keras.layers.Conv1D(32, 3, padding='same', activation='relu')(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Conv1D(64, 3, padding='same', activation='relu')(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.GlobalAveragePooling1D()(x)
    x = tf.keras.layers.Dense(128, activation='relu')(x)
    x = tf.keras.layers.Dropout(0.3)(x)
    x = tf.keras.layers.Dense(num_classes)(x)
    return tf.keras.Model(x_input, x, name="CNNClassifier")

# === Instantiate Models ===
G = build_generator()
D = build_discriminator()
CNN = build_cnn_classifier()

# === Optimizers ===
opt_G = tf.keras.optimizers.Adam(0.0002)
opt_D = tf.keras.optimizers.Adam(0.0002)
opt_CNN = tf.keras.optimizers.Adam(0.001)
loss_fn_CNN = focal_loss()

# === Training Loop ===
epochs = 30
for epoch in range(epochs):
    d_loss_avg, g_loss_avg, c_loss_avg = 0, 0, 0
    step = 0

    for xb, yb in train_dataset:
        step += 1
        bs = tf.shape(xb)[0]
        yb_oh = one_hot(yb)

        # === Generate Fake Data ===
        z = tf.random.normal((bs, latent_dim))
        x_fake = G([z, yb_oh])

        # === Train Discriminator ===
        with tf.GradientTape() as tape_D:
            d_real = D([xb, yb_oh])
            d_fake = D([x_fake, yb_oh])
            loss_D_real = tf.keras.losses.binary_crossentropy(tf.ones_like(d_real) * label_smooth_real, d_real)
            loss_D_fake = tf.keras.losses.binary_crossentropy(tf.zeros_like(d_fake), d_fake)
            d_loss = tf.reduce_mean(loss_D_real + loss_D_fake)

        grads_D = tape_D.gradient(d_loss, D.trainable_variables)
        opt_D.apply_gradients(zip(grads_D, D.trainable_variables))

        # === Train Generator ===
        z = tf.random.normal((bs, latent_dim))
        x_fake = G([z, yb_oh])
        with tf.GradientTape() as tape_G:
            d_fake = D([x_fake, yb_oh])
            g_loss = tf.reduce_mean(tf.keras.losses.binary_crossentropy(tf.ones_like(d_fake), d_fake))

        grads_G = tape_G.gradient(g_loss, G.trainable_variables)
        opt_G.apply_gradients(zip(grads_G, G.trainable_variables))

        # === Train CNN Classifier ===
        x_synthetic = G([tf.random.normal((bs, latent_dim)), yb_oh])
        x_combined = tf.concat([xb, x_synthetic], axis=0)
        y_combined = tf.concat([yb, yb], axis=0)
        with tf.GradientTape() as tape_CNN:
            logits = CNN(x_combined)
            c_loss = loss_fn_CNN(y_combined, logits)

        grads_CNN = tape_CNN.gradient(c_loss, CNN.trainable_variables)
        opt_CNN.apply_gradients(zip(grads_CNN, CNN.trainable_variables))

        d_loss_avg += d_loss.numpy()
        g_loss_avg += g_loss.numpy()
        c_loss_avg += c_loss.numpy()

    print(f"🌀 Epoch {epoch+1:02d} | D Loss: {d_loss_avg/step:.4f} | G Loss: {g_loss_avg/step:.4f} | CNN Loss: {c_loss_avg/step:.4f}")

# === Evaluation ===
y_preds, y_trues = [], []
for xb, yb in test_dataset:
    logits = CNN(xb, training=False)
    y_pred = tf.argmax(logits, axis=1).numpy()
    y_preds.extend(y_pred)
    y_trues.extend(yb.numpy())

print("\n✅ Accuracy:", accuracy_score(y_trues, y_preds))
print("\n📊 Classification Report:\n", classification_report(y_trues, y_preds, target_names=[label_map[i] for i in sorted(label_map)]))

# === Save Models ===
G.save("generator_hybrid.keras")
D.save("discriminator_hybrid.keras")
CNN.save("external_cnn_classifier.keras")
print("\n💾 Models saved as `.keras` format.")
```
